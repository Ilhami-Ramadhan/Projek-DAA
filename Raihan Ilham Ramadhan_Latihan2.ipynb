```python
#Latihan 9
def getKali(numbers):
    hasil = 1
    for num in numbers:
        hasil *= num  # Perkalian tiap elemen
    return hasil

# Contoh list angka
myList = [1, 2, 3, 4]

# Menggunakan fungsi getKali
hasil_kali = getKali(myList)
print("Hasil perkalian seluruh angka:", hasil_kali)
```

```python
#Latihan 10
def getBagi(angka_list):
    if len(angka_list) == 0:
        raise ValueError("Daftar tidak boleh kosong")

    hasil = angka_list[0]
    for angka in angka_list[1:]:
        if angka == 0:
            raise ValueError("Pembagian dengan nol tidak diperbolehkan")
        hasil /= angka
    return hasil

# Contoh penggunaan fungsi
angka_list = [100, 5, 2]  # Misalnya kita ingin 100 dibagi 5 lalu dibagi 2
print("Hasil pembagian:", getBagi(angka_list))
```

```python
#Latihan 11
def getSum(angka_list):
    """Fungsi untuk menjumlahkan seluruh angka dalam daftar."""
    return sum(angka_list)

def getBagi(himpunan1, himpunan2):
    """Fungsi untuk membagi hasil jumlah dua himpunan angka."""
    sum1 = getSum(himpunan1)
    sum2 = getSum(himpunan2)

    if sum2 == 0:
        raise ValueError("Pembagian dengan nol tidak diperbolehkan")

    return sum1 / sum2

# Contoh penggunaan fungsi
himpunan1 = [10, 20, 30]  # Himpunan pertama
himpunan2 = [2, 3]        # Himpunan kedua

print("Hasil pembagian:", getBagi(himpunan1, himpunan2))
```

```python
#Latihan 12
def getSum(angka_list):
    """Fungsi untuk menjumlahkan seluruh angka dalam daftar."""
    return sum(angka_list)

def getKurang(himpunan1, himpunan2):
    """Fungsi untuk mengurangi hasil jumlah dua himpunan angka."""
    sum1 = getSum(himpunan1)
    sum2 = getSum(himpunan2)

    return sum1 - sum2

# Contoh penggunaan fungsi
himpunan1 = [10, 20, 30]  # Himpunan pertama
himpunan2 = [5, 15]       # Himpunan kedua

print("Hasil pengurangan:", getKurang(himpunan1, himpunan2))
```
